```python
import numpy as np
from scipy.stats import kstest
import matplotlib.pyplot as plt
from scipy.stats import norm

# Daily temperature data (Â°C)
temps = [15, 17, 16, 14, 18, 20, 19, 15, 21, 22]

# Calculate the mean and standard deviation using NumPy
mu = np.mean(temps)
sigma = np.std(temps)
z = (temps - mu) / sigma  # % Standardized data

# Perform the K-S test
h, p = kstest(z, 'norm')

# Output the results
print("Test Decision (h):", h)
print("P-value:", p)
print("KS Statistic:", kstest(z, 'norm')[0])


# Plot the empirical CDF
x = np.linspace(min(z), max(z), 100)
plt.plot(x, norm.cdf(x), label="Standard Normal CDF")
plt.hist(z, cumulative=True, density=True, bins=10, label="Empirical CDF")
plt.xlabel('Scaled Temperature')
plt.ylabel('Cumulative Probability')
plt.title('Empirical CDF vs. Standard Normal CDF')
plt.legend()
plt.show()

# Conclusion
alpha = 0.05  # Significance level
if p < alpha:
    print("\nConclusion: The null hypothesis is rejected. The temperature data does not appear to come from a normal distribution.")
else:
    print("\nConclusion: The null hypothesis cannot be rejected. The temperature data may come from a normal distribution.")
```
